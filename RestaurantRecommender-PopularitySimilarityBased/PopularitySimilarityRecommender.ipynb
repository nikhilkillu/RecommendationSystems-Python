#### Choosing Footfall option - List of other restaurants that builds up a queue on first selection ...

```python
userplacerelationship = pd.pivot_table(data=userrating, values='rating', index='userID', columns='placeID')
userplacerelationship.head()
```

```python
#I haven't visited Le Mertidian Delhi, how can irate it zero !
#I visited Shubham Restaurant, the food was pathetic, I rated it zero!
# The difference is clear!

#userplacerelationship.fillna(0, inplace = True )
#userplacerelationship.head()
```

```python
randomPlace = userplacerelationship[135085]
randomPlace[randomPlace>=0]
randomPlace.head()
```

##### Evaluating Similarity Based on Correlation between Target Column versus All columns

```python
columns_similar_to_RandomPlace = userplacerelationship.corrwith(randomPlace)
```

```python
columns_similar_to_RandomPlace.head()
```

```python
print(type(columns_similar_to_RandomPlace))
```

```python
similarPlaces = pd.DataFrame(columns_similar_to_RandomPlace, columns=['PearsonR'])
similarPlaces.dropna(inplace=True)
similarPlaces.head()
```

```python
sortedSimilarPlaces = similarPlaces.sort_values('PearsonR', ascending=False).head(10)
sortedSimilarPlaces.head(20)
```
