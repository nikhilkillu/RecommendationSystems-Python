## Top IMDB Movies!  

```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('C://Users//Lenovo//Desktop//IPY//movies_metadata.csv', low_memory=False)
data.head(2)
```

```python
# Discovering all Columns:
data.columns
```

```python
#IMDB Rating = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C where:

#R = average for the movie (mean) = (Rating)
#v = number of votes for the movie
#m = minimum votes required to be listed
#C = the mean vote across the whole report
```

```python
#Cherry picking columns for IMDB Formula:
data = data[['title','vote_average','vote_count']]
data.head(10)
```

```python
# Calculating C
C = data['vote_average'].mean()
print(C)
# Calculating m
m = data['vote_count'].quantile(0.80)
print(m)
```

![image](https://user-images.githubusercontent.com/689226/79051634-c5708380-7c4e-11ea-837d-89ff8bed08a9.png)


```python
data = data[data['vote_count'] >= m]
data.head(10)
```

```python
def weighted_rating(x):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Define a new feature 'score' and calculate its value with `weighted_rating()`
data['score'] = data.apply(weighted_rating, axis=1)
data.head(10)
```

```python
#Sort movies based on score calculated above
data = data.sort_values('score', ascending=False)

#Print the top 15 movies
data[['title', 'score']].head(10)
```
